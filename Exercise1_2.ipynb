# Exercise

通過創建SinSignal和CosSignal對象並將它們加起來來合成複合信號。 評估信號以獲得Wave，然後聆聽。 計算其頻譜並繪製。 如果添加的頻率分量不是基頻的倍數，會發生什麼？

Solution
這是我選擇的一些任意組件。 它產生了一個有趣的波形！

```python
signal = (thinkdsp.SinSignal(freq=400, amp=1.0) +
          thinkdsp.SinSignal(freq=600, amp=0.5) +
          thinkdsp.SinSignal(freq=800, amp=0.25))
signal.plot()
```

我們可以使用該信號進行波動：

```python
wave2 = signal.make_wave(duration=1)
wave2.apodize()
```

聽起來是這樣的：

```python
wave2.make_audio()
```

這些分量都是200 Hz的倍數，因此它們發出連貫的聲音。

頻譜如下所示：

```python
spectrum = wave2.make_spectrum()
spectrum.plot(high=2000)
```

如果我們添加的分量不是200 Hz的倍數，我們會以不同的音調聽到它。

```python
signal += thinkdsp.SinSignal(freq=450)
signal.make_wave().make_audio()
```
